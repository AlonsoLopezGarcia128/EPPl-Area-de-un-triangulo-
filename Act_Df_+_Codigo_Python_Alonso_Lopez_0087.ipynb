## Formula para calcular Área

```python
# Calcular area de triangulo
# Obtener altura y guardarla en variable H
# Obtener base y guardarla en variable B
# Obtener area con formula A = (B*H)/2
print("Programa para calcular el area de un triangulo")
B = float(input("Ingresa la base: "))
H = float(input("Ingresa la altura: "))
A = (B*H)/2
print("El area del triangulo es: ", A)
print("Programa realizado por Alonso Lopez Garcia NL = 31")
```
